Descriptive Statistics

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from numpy import random
```

```python
samples=np.random.normal(100,15,1000)
samples
```

```python
mean = np.mean(samples)
median = np.median(samples)
mode = stats.mode(samples)
print("mean= ",mean)
print("median= ",median)
print("mode= ",mode)
```

```python
min = np.min(samples)
max = np.max(samples)
q1 = np.percentile(samples, 25)
q3 = np.percentile(samples, 75)
iqr = stats.iqr(samples) # iqr = q3 - q1
print("min= ",min)
print("max= ",max)
print("q1= ",q1)
print("q3= ",q3)
print("iqr= ",iqr)
```

```python
variance = np.var(samples)
std_dev = np.std(samples)
print("variance= ",variance)
print("std_dev= ",std_dev)
```

```python
skewness = stats.skew(samples) #If the skewness is between -0.5 and 0.5 the data are fairy symmetrical
kurtosis = stats.kurtosis(samples) #If the kurtosis is between -1 and 0.5 the data are fairy symmetrical
print("skewness= ",skewness)
print("kurtosisss= ",kurtosis)
```

NumPy Correlation Calculation

```python
x=np.arange(10,20)
y=random.randint(5,100,10)
print("x= ",x)
print("y= ",y)

r,p = stats.pearsonr(x,y)
print("r= ",r)
```

Pandas Correlation Calculation

```python
x = pd.Series(range(10, 20))
y = pd.Series([2, 1, 4, 5, 8, 12, 18, 25, 96, 48])
r,p = stats.pearsonr(x,y)
print("r= ", r)
rho,pval = stats.spearmanr(x,y)
print("rho= ", rho)
```
